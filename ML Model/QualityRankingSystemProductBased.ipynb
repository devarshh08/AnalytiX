```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('E:\CODING\PROGRAMS\PROJ EXHIBITION\SmartphonesDataset.csv', encoding = "ISO-8859-1", engine='python')
```

```python
df1.head()
```

```python
#Removing the comma in price column and convert to integers
df1['Price Rs'] = df1['Price Rs'].str.replace(',', '').astype(float)
df1
```

```python
"""
now we will sort df1 on the basis of price range and create a seperate dataframe for each price range
price ranges:
0 - 9999
10000 - 19999
20000 - 29999
30000 - 39999
39999 - 40000
40000 - 49999
50000 - 59999
"""
```

```python
df_below10k = df1[(df1['Price Rs'] >= 0) & (df1['Price Rs'] <= 9999)]
df_below20k = df1[(df1['Price Rs'] >= 10000) & (df1['Price Rs'] <= 19999)]
df_below30k = df1[(df1['Price Rs'] >= 20000) & (df1['Price Rs'] <= 29999)]
df_below40k = df1[(df1['Price Rs'] >= 30000) & (df1['Price Rs'] <= 39999)]
df_below50k = df1[(df1['Price Rs'] >= 40000) & (df1['Price Rs'] <= 49999)]
df_below60k = df1[(df1['Price Rs'] >= 50000) & (df1['Price Rs'] <= 59999)]
df_below10k
```

```python
#also grouping phones having same names and only having the columns which will actually be useful
df_below10k2 = df_below10k.copy()
condition = df_below10k['Name'].str.split().str[:3].str.join(' ')
df_below10k2['phone_model'] = condition
df_below10k2 = df_below10k2.groupby('phone_model').agg({'Rating' : 'mean', 'Price Rs' : 'mean'}).reset_index()
df_below10k = df_below10k2
df_below10k
```

```python
#also grouping phones having same names and only having the columns which will actually be useful
df_below20k2 = df_below20k.copy()
condition = df_below20k['Name'].str.split().str[:3].str.join(' ')
df_below20k2['phone_model'] = condition
df_below20k2 = df_below20k2.groupby('phone_model').agg({'Rating' : 'mean', 'Price Rs' : 'mean'}).reset_index()
df_below20k = df_below20k2
df_below20k
```

```python
#also grouping phones having same names and only having the columns which will actually be useful
df_below30k2 = df_below30k.copy()
condition = df_below30k['Name'].str.split().str[:3].str.join(' ')
df_below30k2['phone_model'] = condition
df_below30k2 = df_below30k2.groupby('phone_model').agg({'Rating' : 'mean', 'Price Rs' : 'mean'}).reset_index()
df_below30k = df_below30k2
df_below30k
```

```python
#also grouping phones having same names and only having the columns which will actually be useful
df_below40k2 = df_below40k.copy()
condition = df_below40k['Name'].str.split().str[:3].str.join(' ')
df_below40k2['phone_model'] = condition
df_below40k2 = df_below40k2.groupby('phone_model').agg({'Rating' : 'mean', 'Price Rs' : 'mean'}).reset_index()
df_below40k = df_below40k2
df_below40k
```

```python
#also grouping phones having same names and only having the columns which will actually be useful
df_below50k2 = df_below50k.copy()
condition = df_below50k['Name'].str.split().str[:3].str.join(' ')
df_below50k2['phone_model'] = condition
df_below50k2 = df_below50k2.groupby('phone_model').agg({'Rating' : 'mean', 'Price Rs' : 'mean'}).reset_index()
df_below50k = df_below50k2
df_below50k
```

```python
#also grouping phones having same names and only having the columns which will actually be useful
df_below60k2 = df_below60k.copy()
condition = df_below60k['Name'].str.split().str[:3].str.join(' ')
df_below60k2['phone_model'] = condition
df_below60k2 = df_below60k2.groupby('phone_model').agg({'Rating' : 'mean', 'Price Rs' : 'mean'}).reset_index()
df_below60k = df_below60k2
df_below60k
```

```python
#sorting each dataframe

#ranking each dataframe below the given price range based on company i.e sales in millions

df_below10k = df_below10k.sort_values(by='Rating', ascending = False)
df_below20k = df_below20k.sort_values(by='Rating', ascending = False)
df_below30k = df_below30k.sort_values(by='Rating', ascending = False)
df_below40k = df_below40k.sort_values(by='Rating', ascending = False)
df_below50k = df_below50k.sort_values(by='Rating', ascending = False)
df_below60k = df_below60k.sort_values(by='Rating', ascending = False)
```

```python
df_below10k.to_csv('product_based_ranking_below10k.csv', index = False)
df_below20k.to_csv('product_based_ranking_below20k.csv', index = False)
df_below30k.to_csv('product_based_ranking_below30k.csv', index = False)
df_below40k.to_csv('product_based_ranking_below40k.csv', index = False)
df_below50k.to_csv('product_based_ranking_below50k.csv', index = False)
df_below60k.to_csv('product_based_ranking_below60k.csv', index = False)
```

```python
dataframes = [df_below10k, df_below20k, df_below30k, df_below40k, df_below50k, df_below60k]
price_ranges = ['Below 10k', 'Below 20k', 'Below 30k', 'Below 40k', 'Below 50k', 'Below 60k']

for i, df in enumerate(dataframes):
    plt.figure(figsize=(10, 6))
    sns.regplot(x='Price Rs', y='Rating', data=df, scatter_kws={"s": 50}, ci=None)
    plt.xlabel('Price Rs')
    plt.ylabel('Rating')
    plt.title(f'Regression Plot: Price Rs vs. Rating for {price_ranges[i]}')
    plt.grid(True)
    plt.show()
```
